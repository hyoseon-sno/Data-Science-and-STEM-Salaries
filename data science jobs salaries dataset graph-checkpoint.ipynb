```python
#x=salary, y=experience level

import csv
import matplotlib.pyplot as plt



f=open("Data Science Jobs Salaries.csv")

data=csv.reader(f)


salary_en = [] #봉급
salary_mi = []
salary_se = []
salary_ex = []

for row in data:
    if 'EN' in row[1]:
        salary_en.append(row[4])
    if 'MI' in row[1]:
        salary_mi.append(row[4])
    if 'SE' in row[1]:
        salary_se.append(row[4])
    if 'EX' in row[1]:
        salary_ex.append(row[4])


f.close()
##########[시각화]###############
plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('experience_level and salary')

plt.plot(salary_en, color="red", label = "EN")
plt.plot(salary_mi, color="blue", label = "MI")
plt.plot(salary_se, color="green", label = "SE")
plt.plot(salary_ex, color="black", label = "EX")

plt.legend()

plt.show()

```

```python
#x=salary, y= employment_type

import csv
import matplotlib.pyplot as plt



f=open("Data Science Jobs Salaries.csv")

data=csv.reader(f)


salary_pt = [] #봉급
salary_ft = []
salary_ct = []
salary_fl = []

for row in data:
    if 'PT' in row[2]:
        salary_pt.append(row[4])
    if 'FT' in row[2]:
        salary_ft.append(row[4])
    if 'CT' in row[2]:
        salary_ct.append(row[4])
    if 'FL' in row[2]:
        salary_fl.append(row[4])


f.close()
##########[시각화]###############
plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('employment_type and salary')

plt.plot(salary_pt, color="red", label = "PT")
plt.plot(salary_ft, color="blue", label = "FT")
plt.plot(salary_ct, color="green", label = "CT")
plt.plot(salary_fl, color="black", label = "FL")

plt.legend()

plt.show()
```
